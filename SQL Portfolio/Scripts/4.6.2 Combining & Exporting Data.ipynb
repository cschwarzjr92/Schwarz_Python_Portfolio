# Merge df_ords_combined with df_prods

```python
# Merge df_prods and df_ords_combined -- I renamed the indicator as 'flag' because the command would not run otherwise
df_merged = df_ords_combined.merge(df_prods, on = 'product_id', indicator = True)
```
