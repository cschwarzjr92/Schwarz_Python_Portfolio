# Task 3 - Create a Histogram

```python
# Create a histogram of order_hour_of_day; divided into 24 bins for 24 hours

hist1 = instacart['order_hour_of_day'].plot.hist(bins = 24, color = 'teal')
```

# Task 4 - Create a Bar Chart

```python
# Assuming 'instacart' is your DataFrame
bar1 = instacart['loyalty_flag'].value_counts().plot.bar(color=['navy', 'teal', 'lightblue'], edgecolor='black', rot=0)

bar1.set_title('Distribution of Customers by Loyalty Status')

# Disable scientific notation for the y-axis labels
bar1.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.show()
```

# Task 5 - Check Prices vs. Hour of the Day (Line Graph)

```python
# Create a random sample - generate seed
np.random.seed(4)
```

```python
# Create dev list with 70/30 ration

dev = np.random.rand(len(instacart)) <= 0.7
```

```python
# Create 'big' sample

big = instacart[dev]
```

```python
# create 'small' sample

small = instacart[~dev]
```

```python
# Compare size of data frame to confirm accuracy

len(instacart)
```

```python
len(big) + len(small)
```

```python
# Create subset of sample using only needed columns

instacart_2 = small[['order_hour_of_day', 'prices']]
```

```python
# check head of subset

instacart_2.head()
```

```python
# Create line chart of order_hour_of_day and prices

line1 = sns.lineplot(data=instacart_2, x='order_hour_of_day', y='prices')
line1.set_title("Average Price of Products Sold by Hour")
```

# Task 7 - Create a scatterplot

```python
# create subset of sample with only needed columns

instacart_4 = small[['age', 'income']]
```

```python
instacart_4.head()
```

```python
scatterplot = sns.scatterplot(x = 'age', y = 'income', data = instacart_4)
```

There is a wide range of incomes for all age groups, however, there are many more individuals with income of 40,000 once they reach the age 40. This would seem to indicate a general trend in income increasing with age, at least to a certain point. Actual mathmetical analysis would be needed to determine whether the trend observed in this chart is statistically signficant. Generally speaking, though, the chart shows the more people over the age over 40 may be higher earners then people younger than 40.
