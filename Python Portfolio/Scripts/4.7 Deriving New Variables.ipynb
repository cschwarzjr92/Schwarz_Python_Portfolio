# Practice - Creating New Column to Categorize Price Range

```python
# Creating subset of ords_prods_merge
df = ords_prods_merge[:1000000]
```

```python
ords_prods_merge.shape
```

```python
df.shape
```

```python
# Create user defined function to categorize items into low, mid, high and not enough data price ranges

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Run price_label function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check value counts of dataframe subset

df['price_range'].value_counts(dropna = False)
```

```python
# check max value of dataframe subset

df['prices'].max()
```

# Practice - Use loc function to create new price range column in subset

```python
#Use loc function to categorize high range products

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#  Use loc function to categorize medium range products
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
# use loc function to categorize low range products

df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

# Practice - If Statements w/ For-Loops  - Busiest Day Column

```python
# Check order_dow column for busiest day of the week 

ords_prods_merge['order_dow'].value_counts(dropna = False)
```

```python
# Use a for-loop to categorize days of week as busiest, least busy, and regularly busy

result = []

for value in ords_prods_merge["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Check result of previous command

result
```

```python
# Combine result from for-loop with ords_prods_merge into new column

ords_prods_merge['busiest_day'] = result
```

```python
# check value count of busiest day column

ords_prods_merge['busiest_day'].value_counts(dropna=False)
```

```python
# check head with new column

ords_prods_merge.head()
```

# Task 2 - Create a new Busiest Days Column

```python
result_2 = []

for value in ords_prods_merge["order_dow"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

```python
result_2
```

# Task #3 Check Values of Busiest Days Column

```python
ords_prods_merge['busiest_days'] = result_2
```

```python
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

The total value count for the new busiest days column is equal to that of the prior busiest day column. The busiest and slowest days category also adds up to the sum of saturday/sunday orders - this confirms that the values are correct.

# Task #4 Create new "Busiest period of day" column with 'most orders', 'average orders', and 'fewest orders'

```python
# Check value counts for order_hour_of_day column

ords_prods_merge['order_hour_of_day'].value_counts()
```

The 24 hours of the day can be split into equal thirds using these labels
    Most orders = 10, 11, 12, 13, 14, 15, 16
    Fewest Orders = 1, 2, 3, 4, 5, 6, 0, 23
    Average Orders = remaining hours

```python
# Create a for-loop if statement to categorize different hours of the day as 'most orders', 'fewest orders' and 'average orders'

result_3 = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value in [9, 10, 11, 12, 13, 14, 15, 16]:
        result_3.append("Most Orders")
    elif value in [0, 1, 2, 3, 4, 5, 6, 23]:
        result_3.append("Fewest Orders")
    else:
        result_3.append("Average Orders")
```

```python
# Check result_3

result_3
```

```python
# Create new column "busiest_period_of_day" in ords_prods_merge

ords_prods_merge['busiest_period_of_day'] = result_3
```

```python
# Check head

ords_prods_merge.head()
```
